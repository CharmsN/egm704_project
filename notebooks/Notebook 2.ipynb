### LiDAR processing 

### Running process_lidar within egm704_env

Organise LiDAR tiles and AOIs in a consistent project structure.

Use a Python script (process_lidar.py) to:

- clip each 1 m LiDAR tile to the AOI,

- generate a 1 m hillshade per tile,

- resample each tile to 10 m.

Use GDAL command-line tools to mosaic the per-tile 1 m DTMs into a single DTM per site.

LiDAR tiles are 1 m DTMs named by tile code (e.g. SP88sw.tif).

AOIs are stored in aoi_sites.gpkg, each site as a separate layer in British National Grid (EPSG:27700).

LiDAR Processing is run (cd C:\EGM704\data_sets\egm704_project; python scripts\lidar\process_lidar.py)

Outputs are written to data/processed/lidar/

### Mosaicking per-site 1 m DTMs with GDAL

The mosaicking is done outside Python using GDAL command-line tools.

change directory : cd C:\EGM704\data_sets\egm704_project\data\processed\lidar\desborough_bng

Then run gdal command; 

gdalbuildvrt desborough_bng_DTM_1m_clipped.vrt ^  
desborough_bng_SP78se_DTM_1m_clipped.tif ^  
desborough_bng_SP88sw_DTM_1m_clipped.tif

and then 

gdal_translate desborough_bng_DTM_1m_clipped.vrt desborough_bng_DTM_1m_clipped_merged.tif -co COMPRESS=LZW

the result is saved as, for example, desborough_bng_DTM_1m_clipped_merged.tif 

The output is, for each site:   
Per-tile products:
- *_DTM_1m_clipped.tif
- *_hillshade_1m.tif
- *_DTM_10m.tif

Site-wide 1 m DTM:
- `<site>_DTM_1m_clipped_merged.tif`

These merged 1 m DTMs are what you’ll align with Sentinel-2 / Sentinel-1 and use for any terrain-related metrics (slope, aspect, topographic position, etc.).
